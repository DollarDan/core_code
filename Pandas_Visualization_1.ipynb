```python
# Concatinate 

df1 = pd.DataFrame(
    [[1,2,3],
     [4,5,6],
     [7,8,9]]
)

df2 = df1 + 3

df = pd.concat([df1,df2])
df
```

```python
# merge or join
adf = pd.DataFrame( 
    {'x1': ['A','B','C'],
     'X2': [1,2,3]}
)

bdf = pd.DataFrame(
    {'x1': ['A', 'B', 'A'],
     'X3': ['X','Y','Z']}
)
pd.merge(adf,bdf, how='left', on ='x1')
```

```python
pd.merge(adf,bdf, how='right', on ='x1')
```
