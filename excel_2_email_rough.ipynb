```python
# data = pd.read_excel("E:\Python test.xlsx")
data = pd.DataFrame()
# data.sample(3)
```

```python
data[2:4]['Due date'] = '14/05/2022' # manual date change
data[4:7]['Due date'] = '29/05/2022'
data[2:4]
```

```python
# str(data[0:1]['Due date']).split('/')[2]
temp = data[2:3]['Due date']
temp = str(temp).split()[1]
temp
data['new_dates'] = data['Task no']
data['target_15'] = data['Task no']
data['target_30'] = data['Task no']
```

```python
data.head(2)
```

```python
for i in range(len(data)):
    # print(type(str((data[i-1:i]['Due date']))))
    temp = data[i:i+1]['Due date']
    temp = str(temp).split()[1]
    # print(temp)
    temp = datetime.strptime(temp, "%d/%m/%Y").date()
    # data.iloc(i) = temp
    data.iat[i, data.columns.get_loc('new_dates')] = temp
    data.iat[i, data.columns.get_loc('target_15')] = temp - timedelta(days=15)
    data.iat[i, data.columns.get_loc('target_30')] = temp - timedelta(days=30)
    
data.head(3)
```

```python
today = date.today()
target_30_data = data.loc[data['target_30'] == today]
target_30_data
```
